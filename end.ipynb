```python
import pandas as pd, numpy as np
```

```python
#load dataset
df = pd.read_csv('./helping/Chapter_13_End_to_end_example/titanic.csv',index_col="PassengerId")
df.sample(4)
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['Cabin'],axis=1)
df['Age'] = df['Age'].fillna(df['Age'].median)
df['Embarked'] = df['Embarked'].fillna('U')
```

```python
df.columns
gender_column = pd.get_dummies(df['Sex'],prefix='Sex')
embarked_column = pd.get_dummies(df['Embarked'],prefix='Embarked')
```

```python
df = pd.concat([df,gender_column],axis=1)
df = pd.concat([df,embarked_column],axis=1)
df = df.drop(columns=['Sex','Embarked'],axis=1)
```
